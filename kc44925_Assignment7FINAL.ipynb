## Step 1: Data Extraction

The data found here today was collected through a google survey asking the following questions:

    1. What college are you in at UT-Austin?
    2. How many credit hours are you taking this semester?
    3. How many years have you completed at UT?
    4. How many hours of sleep do you get on average?
    5. On a scale of 1-10 how happy are you right now? (1 = lowest)
    
50+ responses were collected from the survey

```python
import pandas as pd
import numpy as np
import scipy
from matplotlib import pyplot as plt
```

```python
df = pandas.read_csv("raw_student_satisfaction_survey.csv")
df = pandas.DataFrame(df)

df.head()
```

## Step 2: Data Transformation and Load

Processing of the data consited of deleting the "Timestap" column, all other data was clean enough for processing

```python
final_df = df.drop("Timestamp", axis='columns')
# display(final_df)
final_df.head()
```

```python
final_df.to_csv("student_satisfaction_survey.csv")
print("Data Succesfully updated to CSV...")
```

## Step 3: Analysis: Summary and visualizations

Today, I will be comparing:

    1. A student's average amount of sleep with their happiness level
    2. A student's College/School at UT versus their sleep at UT
    3. Frequency of college mentioned in the survey

### Frequency of College/School

```python
column_names = final_df.columns
print(column_names)

specific_ut_college = final_df["specific_ut_college"]
# print(specific_ut_college)
specific_ut_college.hist(bins=8)

plt.xticks(rotation=45, ha='right')
plt.xlabel("College/School attended within UT-Austin")
plt.ylabel("Frequency")

plt.title("Frequency of specific college/school attendance at UT-Austin")
```

### Histogram Conclusion
The college/school most frequently put in this survey was the College of Natural Science

```python
sleep_hours = final_df["sleep_hours"]
happiness_scale = final_df["happiness_scale"]

plt.bar(sleep_hours, happiness_scale)
plt.xlabel("Average amount of sleep (hours)")
plt.ylabel("Average happiness (1=lowest)")
```

```python
plt.boxplot(sleep_hours)
sleep_hours.describe()

plt.xlabel("Sleep")
plt.ylabel("Sleep in Hours")
```

```python
plt.boxplot(happiness_scale)
happiness_scale.describe()

plt.xlabel("Happiness")
plt.ylabel("Happiness (Scale 1-10)")
plt.title("Happiness Scale")
```

```python
from scipy.stats import pearsonr

p = pearsonr(final_df['happiness_scale'],final_df["sleep_hours"])
print(p[0])
```

### Comparison between Happiness and Sleep Hours Conclusion

As shown from the descriptive statistics, the average happiness for this set of students at UT is around a 7 out of 10 with the lowest being 3. Further data collection would be need to determine the criteria for happiness

On average, UT students sleep 6.77 hours a day with the lowest amount of sleep being 4 hours.

Based on the scatter plot, I can conclude that there is a loose relation between sleep and happiness. When looking at the correlation coeffecient it is a weak association
